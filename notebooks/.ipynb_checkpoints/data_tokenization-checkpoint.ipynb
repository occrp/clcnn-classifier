```python
import pandas as pd
import constants as const
from custom_encoder import Encoder
from sklearn.utils import shuffle

df = pd.read_pickle('../prepared_data/companies_people_fortified.pkl')
encoder = Encoder(const.ALPHABET)
```

```python
df
```

```python
df[0] = encoder.tokenize(df[0])
```

```python
df
```

```python
df = shuffle(df).reset_index(drop=True)
```

```python
df
```

```python
# Prepare indeces to split dataset
index_60 = int(df.shape[0]*0.6)
index_90 = int(df.shape[0]*0.9)
```

```python
# Split dataframe into training(60%), test(30%), validation(10%) 
train_df = df.iloc[0:index_60]
test_df = df.iloc[index_60:index_90]
validation_df = df.iloc[index_90:df.shape[0]]
```

```python
train_df.to_pickle('../prepared_data/train_df_f.pkl')
test_df.to_pickle('../prepared_data/test_df_f.pkl')
validation_df.to_pickle('../prepared_data/validation_df_f.pkl')
```
