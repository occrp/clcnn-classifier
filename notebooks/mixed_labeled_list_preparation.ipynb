```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
```

```python
companies_file = '../data/companies_fortified.csv'
people_file = '../data/people.csv'

companies = pd.read_csv(companies_file, header=None, error_bad_lines=False)
people = pd.read_csv(people_file, header=None, error_bad_lines=False)
```

```python
companies
```

```python
people
```

```python
companies.drop_duplicates(inplace=True)
people.drop_duplicates(inplace=True)
companies = shuffle(companies).reset_index(drop=True)
companies = companies[:people.size] # Make number of companies and people equal
people = shuffle(people).reset_index(drop=True)
```

```python
companies[1] = 0
people[1] = 1
```

```python
companies
```

```python
people
```

```python
companies_people = pd.concat([companies, people])
companies_people = shuffle(companies_people).reset_index(drop=True)
```

```python
companies_people
```

```python
companies_people.to_pickle('../prepared_data/companies_people_fortified.pkl')
```
